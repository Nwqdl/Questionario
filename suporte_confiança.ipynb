```python
import pandas as pd
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
dados = pd.read_csv('form - form.csv') #lendo o csv
#calculando o suporte
suportes = apriori(dados, min_support=0.5, use_colnames=True)
suportes #imprime os suportes
```

```python
regras = association_rules(suportes, metric="confidence", min_threshold=0.9)
regras
```
